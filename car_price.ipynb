```python
df['Car_Age'] = 2025 - df["Year"]
df.drop(columns = "Year",inplace = True)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
sc = StandardScaler()
```

```python
df[['Selling_Price','Present_Price','Driven_kms']] = sc.fit_transform(df[['Selling_Price','Present_Price','Driven_kms']])
```

```python
x = df.drop(columns = "Present_Price")
y = df['Present_Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 48)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pre = lr.predict(x_test)
```

```python
print("MSE: ",mean_squared_error(y_test, y_pre))
print("r2_score: ",r2_score(y_test,y_pre ))
```
